ت) تابع جرم احتمالی که در مورد (ب) به دست آوردید را به ازای $n = 1000$ و  $-1000 < z < 1000$ رسم کنید. نمودار نهایی شبیه کدام توزیع شده است؟

```r
p_Zn <- function(n, z, p) {
  k_plus <- (n + z) / 2
  k_minus <- (n - z) / 2
  if (k_plus %% 1 == 0 && k_minus %% 1 == 0 && k_plus >= 0 && k_minus >= 0) {
    prob <- choose(n, k_plus) * p^k_plus * (1 - p)^k_minus
  } else {
    prob <- 0
  }
  return(prob)
}

n <- 1000
p <- 0.5
z_values <- -1000:1000
probabilities <- sapply(z_values, function(z) p_Zn(n, z, p))

plot(z_values, probabilities, type = "h", lwd = 2, col = "blue", 
     xlab = "z", ylab = "P[Z_n = z]", main = "Exact Distribution of Z_n for n=1000")

axis_x <- seq(-1000, 1000, by = 500)
axis_y <- seq(0, max(probabilities), by = 0.005) 


abline(v = axis_x, col = "gray", lty = "dotted")

abline(h = axis_y, col = "gray", lty = "dotted")
```

```r
#todo
N <- 10000
n_steps <- 1000
target_station <- 2
visited_target <- 0

calculate_prob <- function(n_steps, target_station) {
  position <- 0
  for (j in 1:n_steps) {
    move <- sample(c(-1, 1), 1, prob = c(1 - 0.3, 0.3))
    position <- position + move
    if (position == target_station) {
      return (TRUE)
    }
  }
  return (FALSE)
}
visited_target <- replicate(N, calculate_prob(n_steps, target_station))

estimated_probability <- mean(visited_target)
cat("Estimated probability of visiting x = 2 in a 1000-step random walk:", estimated_probability, "\n")
```

```r
N <- 10000
n_steps_max <- 1000
p <- 0.55
target_stations <- 1:50

calculate_time_to_reach <- function(n_steps_max, target_station, p) {
  position <- 0
  steps <- 0
  while (position != target_station) {
    move <- sample(c(-1, 1), 1, prob = c(1 - p, p))
    position <- position + move
    steps <- steps + 1
    if (steps > n_steps_max) {
      return(NA)  # اگر بیش از n_steps_max طول کشید، نشان می‌دهد که به ایستگاه نرسیده
    }
  }
  return(steps)
}

time_to_reach <- sapply(target_stations, function(target_station) {
  steps_list <- replicate(N, calculate_time_to_reach(n_steps_max, target_station, p))
  mean(steps_list, na.rm = TRUE)  # میانگین زمان برای رسیدن به هر ایستگاه
})

prob_df <- data.frame(Station = target_stations, Time = time_to_reach)

library(ggplot2)
ggplot(prob_df, aes(x = Station, y = Time)) +
  geom_line(color = "blue") +
  labs(title = "Mean steps to reach each k", x = "Station", y = "Mean number of steps") +
  theme_minimal()
```

```r
N <- 10000 
Z0 <- 100   
target <- 105  
n_values <- c(10, 20, 50, 100, 200) 
simulate_taxi <- function(n_steps, target_position) {
  X <- rnorm(n_steps, mean = 0, sd = 1) 
  Z <- Z0 + cumsum(X) 
  return(sum(Z > target_position) > 0)  
}


results <- sapply(n_values, function(n) {
  hits_target <- replicate(N, simulate_taxi(n, target))
  mean(hits_target)  
})
output <- data.frame(
  Steps = n_values,
  Probability = results
)

print(output)
library(ggplot2)
ggplot(output, aes(x = Steps, y = Probability)) +
  geom_line(color = "blue") + 
  geom_point(color = "red") +  
  labs(title = "Probability of Reaching Target (105) for Different Steps",
       x = "Number of Steps",
       y = "Probability of Reaching 105") +
  theme_minimal() +
  scale_y_continuous(limits = c(0, 1))
```
